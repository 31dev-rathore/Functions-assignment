Q1. What is the difference between a function and a method in Python?

  In Python, a function is a standalone block of code that performs a specific task and is not tied to any object. It is defined using the `def` keyword and called by its name. A method, on the other hand, is a function that is associated with an object and is defined within a class. Methods are called on objects and can access or modify the object's data using the `self` parameter. In short, all methods are functions, but not all functions are methods.

```python

def my_function():
    return "I am a function"

class MyClass:
    def my_method(self):
        return "I am a method"

# Example:
print(my_function())         # Calls function
obj = MyClass()
print(obj.my_method())       # Calls method
```

Q6. Explain the concept of generators in Python and how they are defined.

  Generators are defined using functions but use 'yield' instead of 'return' to produce a series of values lazily.

```python
def counter():
    for i in range(3):
        yield i

# Example:
for number in counter():
    print(number)
```

Q7. What are the advantages of using generators over regular functions?

  Generators offer lazy evaluation, memory efficiency, and improved performance for large data sets.

  

```python
# Example (counter generator):
def counter():
    for i in range(3):
        yield i

for number in counter():
    print(number)
```

Q10. What is the difference between map(), reduce(), and filter() functions in Python?
   
  * map() applies a function to all items in an input list.

  *reduce() applies a function cumulatively to items of an iterable, reducing it to a single value (requires import from functools).

  *filter() filters items out of an iterable based on a condition.

```python
from functools import reduce

# Example:
print(list(map(lambda x: x * 2, [1, 2, 3])))
print(reduce(lambda x, y: x + y, [1, 2, 3, 4]))
print(list(filter(lambda x: x % 2 != 0, [1, 2, 3, 4])))
```

Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
print(sum_of_evens([1, 2, 3, 4, 5, 6]))
```

Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage for numbers 1 to 200:
primes = [n for n in range(1, 201) if is_prime(n)]
print(primes)
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.index = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.n_terms:
            raise StopIteration
        if self.index == 0:
            self.index += 1
            return self.a
        else:
            self.a, self.b = self.b, self.a + self.b
            self.index += 1
            return self.a

# Example usage:
fib = Fibonacci(10)
print(list(fib))  # Output: First 10 Fibonacci numbers
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_2(5):
    print(power)
```

Q7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def file_line_reader(filename):
    """Generator that yields each line from a file."""
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()  # Remove leading/trailing whitespace
    except FileNotFoundError:
        print("File not found.")

# Example usage
file_path = 'sample.txt'  # Replace with your file name

for line in file_line_reader(file_path):
    print(line)
```

Q9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_list = [0, 20, 37, 100]
fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))
print(fahrenheit_list)
```

Q10. Create a Python program that uses filter() to remove all the vowels from a given string.

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

# Example usage:
print(remove_vowels("This is a test."))
```
